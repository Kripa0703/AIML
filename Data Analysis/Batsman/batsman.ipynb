


Data Aggregation & Grouped Analysis
Here, we group the dataset based on different features to better understand patterns and relationships between variables.

```python
#total runs scored against each Opposition
df.groupby(['Opposition'])['RunsScored'].sum()  
```

```python
df.groupby(['Opposition'])['RunsScored'].sum().sort_values()
```

```python
df.groupby(['Opposition'])['RunsScored'].sum().sort_values(ascending=False)
```

```python
df.groupby(['Opposition'])['RunsScored'].mean().sort_values()
```

```python
df.groupby(['Opposition'])['RunsScored'].median().sort_values()
```

```python
df.groupby(['Year','Opposition']).RunsScored.sum().sort_values(ascending=False)
```

```python
df.groupby(['Year','Opposition']).RunsScored.sum().sort_values(ascending=False).head(1)
```

```python
df.groupby(['Opposition'])['RunsScored'].agg(['median', 'sum']) #aggregate
```

```python
df.groupby(['Inning'])['RunsScored'].sum().sort_values()
```

```python
pd.crosstab(index = df['Inning'], columns=df['Dismissal'], values= df['RunsScored'], margins = True, aggfunc='count')
```

```python
df[df['Opposition'] == 'Australia']['RunsScored'].sum()
```

```python
pd.crosstab(index = df['Inning'], columns=df['Opposition'], values= df['RunsScored'], margins = True, aggfunc='count')
```

```python
df['Post_Transform']=df['Year']>2012
```
